```python
import matplotlib.pyplot as plt
plt.scatter(df['timestamp'], df['processId'], color='green')
plt.suptitle('Graph of Timestamp vs ProcessId')
plt.xlabel('Timestamp')
plt.ylabel('Process ID')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(df['timestamp'],df['processId'],color='green')
plt.suptitle('Graph of Timestamp vs ProcessId')
plt.xlabel('Timestamp')
plt.ylabel('Process ID')
plt.show()
```

```python
df_class = df.groupby('eventName').sum(numeric_only=True)
```

```python
df_class
```

```python
df_class
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df_class['processId'], labels=df_class.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Process ID by Event Name')
plt.axis('equal')
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.scatter(df['timestamp'], df['processId'], color='green')
plt.title('Scatter Plot: Timestamp vs Process ID')
plt.xlabel('Timestamp')
plt.ylabel('Process ID')

plt.subplot(1, 2, 2)
top = df_class.head(11).iloc[1:]
plt.pie(top['processId'], labels=top.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Process ID by Event Name (Skipping first record, showing next 10)')
plt.axis('equal')

plt.tight_layout()
plt.savefig('dashboard.png')
```
